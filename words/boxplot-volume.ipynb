```python
import csv
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
import textstat
```

```python
df = pd.read_csv('/Users/carina/Downloads/courses/final thesis/precessed data/ex3/resultvolume/GPT2_allwords.csv')
df1 = pd.read_csv('/Users/carina/Downloads/courses/final thesis/precessed data/ex3/resultvolume/BERT_allwords.csv')

df = df.drop(df.columns[0],axis=1)
df1 = df1.drop(df1.columns[0],axis=1)
heads = df.columns.values.tolist()    # 列名称

source = ['NLP paper','Literature','News','Encyclopedias','Social Media','Speeches','Interview','Instruction Manuals']

source_name = []
for i in range(len(source)):
    temsouce = [source[i]] * 60
    source_name = source_name + temsouce
```

# normalized results of volume

```python
volume_data = pd.DataFrame()
tool_name = []
LM1= ['GPT2'] * 480
LM2= ['BERT'] * 480
LM = LM1 + LM2
LM_name = []
value = []
value0 = []
for head in heads:
    value_GPT2 = []
    value_BERT = []
    tem_value_GPT2 = []
    tem_value_GPT2 = df[head].tolist()
    tem_value_BERT = []
    tem_value_BERT = df1[head].tolist()
    for item in tem_value_GPT2:
        value_GPT2.append(item)
        tool_name.append(head)
    for item in tem_value_BERT:
        value_BERT.append(item)
        tool_name.append(head)
    LM_name = LM_name + LM
    value0 = value0 + value_GPT2 + value_BERT
    value = (value0 - np.mean(value0)) / np.std(value0)

volume_data['volume'] = value
volume_data['WATs'] = tool_name
volume_data['LM'] = LM_name

sns.boxplot(x="WATs", y="volume",hue="LM",data=volume_data,palette="Set3",width=0.5, linewidth=1.0) 
plt.legend(loc = 2, bbox_to_anchor = (1,1))
plt.figure(figsize=(1,1))  

```

# results of volume

```python
volume_data = pd.DataFrame()
tool_name = []
LM1= ['GPT2'] * 480
LM2= ['BERT'] * 480
LM = LM1 + LM2
LM_name = []
value = []
value0 = []
for head in heads:
    value_GPT2 = []
    value_BERT = []
    tem_value_GPT2 = []
    tem_value_GPT2 = df[head].tolist()
    tem_value_BERT = []
    tem_value_BERT = df1[head].tolist()
    for item in tem_value_GPT2:
        value_GPT2.append(item)
        tool_name.append(head)
    for item in tem_value_BERT:
        value_BERT.append(item)
        tool_name.append(head)
    LM_name = LM_name + LM
    value0 = value0 + value_GPT2 + value_BERT

volume_data['volume'] = value0
volume_data['WATs'] = tool_name
volume_data['LM'] = LM_name

sns.boxplot(x="WATs", y="volume",hue="LM",data=volume_data,palette="Set3",width=0.5, linewidth=1.0) 
plt.legend(loc = 2, bbox_to_anchor = (1,1))
plt.figure(figsize=(1,1))  

```

# normalized results of volume-LM dependent

```python
volume_data_GPT2 = pd.DataFrame()
volume_data_BERT = pd.DataFrame()
tool_name_GPT2 = []
tool_name_BERT = []
final_value_GPT2 = []
final_value_BERT = []
for head in heads:
    value_GPT2 = []
    value_BERT = []
    tem_value_GPT2 = []
    tem_value_GPT2 = df[head].tolist()
    tem_value_BERT = []
    tem_value_BERT = df1[head].tolist()
    for item in tem_value_GPT2:
        value_GPT2.append(item)
        tool_name_GPT2.append(head)
    for item in tem_value_BERT:
        value_BERT.append(item)
        tool_name_BERT.append(head)
    final_value_GPT2 = final_value_GPT2 + value_GPT2
    final_value_GPT2_ = (final_value_GPT2 - np.mean(final_value_GPT2)) / np.std(final_value_GPT2)
    final_value_BERT = final_value_BERT + value_BERT
    final_value_BERT_ = (final_value_BERT - np.mean(final_value_BERT)) / np.std(final_value_BERT)
    
volume_data_GPT2['volume'] = final_value_GPT2_
volume_data_GPT2['WATs'] = tool_name_GPT2
volume_data_BERT['volume'] = final_value_BERT_
volume_data_BERT['WATs'] = tool_name_BERT

f, axes = plt.subplots(2, 1,sharex=True, figsize=(22,25))

a = sns.violinplot(x="WATs", y="volume",data=volume_data_GPT2,palette="Set3",width=0.5, linewidth=1.0,ax=axes[0])
b = sns.violinplot(x="WATs", y="volume",data=volume_data_BERT,palette="Set3",width=0.5, linewidth=1.0,ax=axes[1])
```
