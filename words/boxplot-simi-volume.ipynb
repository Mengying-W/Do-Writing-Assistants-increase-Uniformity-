```python
import csv
import pandas as pd
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt 
import textstat
import os
```

# volume result - all words + keywords catrgory

```python
base = '/Users/carina/Downloads/courses/final thesis/precessed data/ex3/avg_simi/euc/'

overall_value_GPT2 = pd.DataFrame()
overall_value_BERT = pd.DataFrame()

overall_value_value_BERT = []
overall_value_tool_BERT = []
overall_value_category_BERT = []
overall_value_value_GPT2 = []
overall_value_tool_GPT2 = []
overall_value_category_GPT2 = []
for root, ds, fs in os.walk(base):
        for f in fs:
            if 'GPT2' in f:
                fullpath = os.path.join(root, f)
                df = pd.read_csv(fullpath)
                df = df.drop(df.columns[0],axis=1)
                #WATs = ['Rephrase','Grammarly','ChatGPT','Wordtune','Quillbot']
                WATs = ['Paragraph','Rephrase','Grammarly','ChatGPT','Wordtune','Quillbot']
                df.columns = WATs
                heads = df.columns.values.tolist()
                
                category = f.split('.')[0].split('_')[-1]
            
                for head in heads:
                    tem = df[head].tolist()
                    for item in tem:
                        overall_value_value_GPT2.append(item)
                        overall_value_tool_GPT2.append(head)
                        overall_value_category_GPT2.append(category)
                    
            if 'BERT' in f:
                fullpath = os.path.join(root, f)
                df = pd.read_csv(fullpath)
                df = df.drop(df.columns[0],axis=1)
                #WATs = ['Rephrase','Grammarly','ChatGPT','Wordtune','Quillbot']
                WATs = ['Paragraph','Rephrase','Grammarly','ChatGPT','Wordtune','Quillbot']
                df.columns = WATs
                heads = df.columns.values.tolist()
                
                category = f.split('.')[0].split('_')[-1]
                for head in heads:
                    tem = df[head].tolist()
                    for item in tem:
                        overall_value_value_BERT.append(item)
                        overall_value_tool_BERT.append(head)
                        overall_value_category_BERT.append(category)
                        
overall_value_GPT2['volume_GPT2'] = overall_value_value_GPT2
overall_value_GPT2['WATs'] = overall_value_tool_GPT2
overall_value_GPT2['category'] = overall_value_category_GPT2

overall_value_BERT['volume_BERT'] = overall_value_value_BERT
overall_value_BERT['WATs'] = overall_value_tool_BERT
overall_value_BERT['category'] = overall_value_category_BERT
```

```python
f, axes = plt.subplots(1, 1,sharex=True, figsize=(22,10))

orders = ['tfidf5','YAKE5','tfidf10','YAKE10','tfidf15','YAKE15','allwords']
#my_pal = {"BERT": "g", "GPT2": "b"}

a = sns.boxplot(y="volume_GPT2",x= "WATs",hue="category", hue_order=orders,data=overall_value_GPT2)
a.legend (bbox_to_anchor = (1.13, 1),fontsize='xx-large')

plt.xlabel("", fontdict={'family' : 'Times New Roman', 'size'   : 320})
plt.ylabel("Euc_GPT2", fontdict={'family' : 'Times New Roman', 'size'   : 32})
plt.yticks(fontproperties = 'Times New Roman', size = 35)
plt.xticks(fontproperties = 'Times New Roman', size = 30)
plt.savefig('/Users/carina/Downloads/courses/final thesis/figures of results/word/average simi/central average euclidean distance GPT2.pdf', format = 'pdf',dpi = 500,bbox_inches='tight')
'''
b = sns.boxplot(y="volume_BERT",x= "WATs",hue="category", hue_order=orders,data=overall_value_BERT)
b.legend (bbox_to_anchor = (1.13, 1),fontsize='xx-large')

plt.xlabel("", fontdict={'family' : 'Times New Roman', 'size'   : 320})
plt.ylabel("Euc_BERT", fontdict={'family' : 'Times New Roman', 'size'   : 32})
plt.yticks(fontproperties = 'Times New Roman', size = 35)
plt.xticks(fontproperties = 'Times New Roman', size = 30)
plt.savefig('/Users/carina/Downloads/courses/final thesis/figures of results/word/average simi/central average euclidean distance BERT.pdf', format = 'pdf',dpi = 500,bbox_inches='tight')
'''
```

```python
base = '/Users/carina/Downloads/courses/final thesis/precessed data/ex3/central similarity/cos/'

overall_value_GPT2 = pd.DataFrame()
overall_value_BERT = pd.DataFrame()

overall_value_value_BERT = []
overall_value_tool_BERT = []
overall_value_category_BERT = []
overall_value_value_GPT2 = []
overall_value_tool_GPT2 = []
overall_value_category_GPT2 = []
for root, ds, fs in os.walk(base):
        for f in fs:
            if 'GPT2' in f:
                fullpath = os.path.join(root, f)
                df = pd.read_csv(fullpath)
                df = df.drop(df.columns[0],axis=1)
                heads = df.columns.values.tolist()
                
                category = f.split('.')[0].split('_')[-1]
            
                for head in heads:
                    tem = df[head].tolist()
                    for item in tem:
                        overall_value_value_GPT2.append(item)
                        overall_value_tool_GPT2.append(head)
                        overall_value_category_GPT2.append(category)
                    
            if 'BERT' in f:
                fullpath = os.path.join(root, f)
                df = pd.read_csv(fullpath)
                df = df.drop(df.columns[0],axis=1)
                heads = df.columns.values.tolist()
                
                category = f.split('.')[0].split('_')[-1]
                for head in heads:
                    tem = df[head].tolist()
                    for item in tem:
                        overall_value_value_BERT.append(item)
                        overall_value_tool_BERT.append(head)
                        overall_value_category_BERT.append(category)
                        
overall_value_GPT2['cos_similarity_GPT2'] = overall_value_value_GPT2
overall_value_GPT2['WATs'] = overall_value_tool_GPT2
overall_value_GPT2['category'] = overall_value_category_GPT2

overall_value_BERT['cos_similarity_BERT'] = overall_value_value_BERT
overall_value_BERT['WATs'] = overall_value_tool_BERT
overall_value_BERT['category'] = overall_value_category_BERT
```

```python
f, axes = plt.subplots(2, 1,sharex=True, figsize=(22,25))

orders = ['TFIDF5','YAKE5','TFIDF10','YAKE10','TFIDF15','YAKE15','allwords']
#my_pal = {"BERT": "g", "GPT2": "b"}

a = sns.boxplot(y="cos_similarity_GPT2",x= "WATs",hue="category", hue_order=orders,data=overall_value_GPT2,ax=axes[0])
a.legend (bbox_to_anchor = (1.1, 1),fontsize='x-large')

b = sns.boxplot(y="cos_similarity_BERT",x= "WATs",hue="category", hue_order=orders,data=overall_value_BERT,ax=axes[1])
b.legend (bbox_to_anchor = (1.1, 1),fontsize='x-large')
```
